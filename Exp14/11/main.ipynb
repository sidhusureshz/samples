```python
from graphics.rectangle import *
from graphics.circle import *
from graphics.graphics3d.cuboid import *
from graphics.graphics3d.sphere import *

while True:
    print("\n1. Rectangle\n2. Circle\n3. Cuboid\n4. Sphere\n5. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        l = float(input("Enter the length of rectangle: "))
        b = float(input("Enter the breadth of rectangle: "))
        print("Area of rectangle:", area(l, b))
        print("Perimeter of rectangle:", perimeter(l, b))

    elif choice == 2:
        r = float(input("Enter the radius of circle: "))
        print("Area of circle:", round(area(r), 2))
        print("Perimeter of circle:", round(perimeter(r), 2))

    elif choice == 3:
        l = float(input("Enter the length of cuboid: "))
        b = float(input("Enter the breadth of cuboid: "))
        h = float(input("Enter the height of cuboid: "))
        print("area of cuboid:", round(area(l, b, h), 2))
        print("Perimeter of cuboid:", round(perimeter(l, b, h), 2))

    elif choice == 4:
        r = float(input("Enter the radius of sphere: "))
        print("Surface area of sphere:", round(area(r), 2))
        print("Volume of sphere:", round(perimeter(r), 2))

    elif choice == 5:
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Please try again.")
```
